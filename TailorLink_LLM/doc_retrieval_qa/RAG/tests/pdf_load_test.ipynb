```python
from pdfminer.high_level import extract_pages
from pdfminer.layout import LTTextBoxHorizontal, LTFigure
import math

file_path = '../../data/documents/genesis-dh-manual-ko.pdf'

def pdf_load(file_path, x_tolerance=100):
    # x: x좌표 값 (정렬 대상인 x값)
    # x_tolerance : 허용 오차 범위 (오차 허용 정돌르 조정하는 값)
    def x_group_key(x, x_tolerance=100): 
        return min(
            [group for group in range(0, 1000, x_tolerance) if math.isclose(x, group, abs_tol=x_tolerance)],
            default=x
        )
    
    contents = []
    for page, page_layout in enumerate(extract_pages(file_path)):    
    
        sorted_objs = sorted(
            (obj for obj in page_layout._objs ),
            key=lambda obj: (x_group_key(obj.bbox[0]), -obj.bbox[3])
        )
    
        text_combine=""
        imgs = []
        texts = []
        for element in sorted_objs:
            if isinstance(element, LTTextBoxHorizontal):
                text_combine += element.get_text()
                text = {
                    "bbox" : element.bbox,
                    "width" : element.width,
                    "height" : element.height,                
                    "x0" : element.x0,
                    "x1" : element.x1,
                    "y0" : element.y0,
                    "y1" : element.y1,
                }
                texts.append(text)
            elif isinstance(element, LTFigure):
                img = {
                    "bbox" : element.bbox,
                    "matrix" : element.matrix,
                    "width" : element.width,
                    "height" : element.height,                
                    "x0" : element.x0,
                    "x1" : element.x1,
                    "y0" : element.y0,
                    "y1" : element.y1,
                }
                imgs.append(img)
        
        content = {
            "page" : page,
            "text_combine" : text_combine,
            "texts" : texts,
            "img" : imgs,
        }
        contents.append(content)
    return contents
```

```python
from pdfminer.high_level import extract_pages
from pdfminer.layout import LTTextBoxHorizontal, LTFigure
from pdfminer.pdfparser import PDFParser
from pdfminer.pdfdocument import PDFDocument
import math
import os

file_path = '../../data/documents/genesis-dh-manual-ko.pdf'

def pdf_load(file_path):
    # x: x좌표 값 (정렬 대상인 x값)
    # x_tolerance : 허용 오차 범위 (오차 허용 정돌르 조정하는 값)
    x_tolerance = 100
    def x_group_key(x): 
        return min(
            [group for group in range(0, 1000, x_tolerance) if math.isclose(x, group, abs_tol=x_tolerance)],
            default=x
        )
    
    # 파일명 추출
    file_name = os.path.splitext(os.path.basename(file_path))[0]
    
    # 메타데이터 추출
    metadata = {"car_model": file_name}
    with open(file_path, 'rb') as f:
        parser = PDFParser(f)
        document = PDFDocument(parser)
        if document.info:
            metadata.update({
                key: value.decode('utf-8', errors='ignore') if isinstance(value, bytes) else value
                for key, value in document.info[0].items()
            })
    
    contents = []
    for page, page_layout in enumerate(extract_pages(file_path)):    
    
        sorted_objs = sorted(
            (obj for obj in page_layout._objs ),
            key=lambda obj: (x_group_key(obj.bbox[0]), -obj.bbox[3])
        )
    
        text_combine = ""
        imgs = []
        texts = []
        for element in sorted_objs:
            if isinstance(element, LTTextBoxHorizontal):
                text_combine += element.get_text()
                text = {
                    "bbox" : element.bbox,
                    "width" : element.width,
                    "height" : element.height,                
                    "x0" : element.x0,
                    "x1" : element.x1,
                    "y0" : element.y0,
                    "y1" : element.y1,
                }
                texts.append(text)
            elif isinstance(element, LTFigure):
                img = {
                    "bbox" : element.bbox,
                    "matrix" : element.matrix,
                    "width" : element.width,
                    "height" : element.height,                
                    "x0" : element.x0,
                    "x1" : element.x1,
                    "y0" : element.y0,
                    "y1" : element.y1,
                }
                imgs.append(img)
        
        content = {
            "page" : page,
            "text_combine" : text_combine,
            "texts" : texts,
            "img" : imgs,
        }
        contents.append(content)
    
    # 메타데이터와 콘텐츠 반환
    return {"metadata": metadata, "contents": contents}

# 실행 예제
result = pdf_load(file_path)
print(result["metadata"])  # 메타데이터 출력
```

```python
import fitz  # PyMuPDF
import cv2
import numpy as np
import os

def extract_and_fix_colors(pdf_path, output_dir, image_format="png"):
    """
    Extract images from a PDF and fix color issues by converting to RGB.
    """
    os.makedirs(output_dir, exist_ok=True)
    pdf_document = fitz.open(pdf_path)

    for page_number in range(len(pdf_document)):
        page = pdf_document[page_number]
        print(f"Processing page {page_number + 1}")

        # Get images on the page
        images = page.get_images(full=True)

        for img_index, img in enumerate(images):
            xref = img[0]
            try:
                # Extract image bytes
                base_image = pdf_document.extract_image(xref)
                image_bytes = base_image["image"]
                
                # Decode the image bytes using OpenCV
                image_array = np.frombuffer(image_bytes, dtype=np.uint8)
                image = cv2.imdecode(image_array, cv2.IMREAD_UNCHANGED)  # Load as is

                # Convert to RGB if the image is not already in RGB format
                if len(image.shape) == 2:  # Grayscale
                    image = cv2.cvtColor(image, cv2.COLOR_GRAY2RGB)
                elif image.shape[2] == 4:  # RGBA (with alpha channel)
                    image = cv2.cvtColor(image, cv2.COLOR_BGRA2BGR)
                elif image.shape[2] == 3:  # Already RGB
                    pass  # No conversion needed

                # Save the image
                output_path = os.path.join(output_dir, f"page{page_number+1}_img{img_index+1}.{image_format}")
                cv2.imwrite(output_path, image)
                print(f"Saved fixed image: {output_path}")
            except Exception as e:
                print(f"Error processing image on page {page_number + 1}, image {img_index + 1}: {e}")

# Example usage
pdf_path = '../../data/documents/genesis-dh-manual-ko.pdf'
output_dir = "../../data/documents/output_images"  # Directory to save extracted images
image_format = "png"  # Format to save images (e.g., "png" or "jpg")

extract_and_fix_colors(pdf_path, output_dir, image_format)
```

```python
from PyPDF2 import PdfReader

def extract_images_with_pypdf2(pdf_path, output_dir):
    """
    Extract images using PyPDF2 to handle Separation color space.
    """
    os.makedirs(output_dir, exist_ok=True)
    reader = PdfReader(pdf_path)

    for page_number, page in enumerate(reader.pages):
        for x_object in page["/Resources"]["/XObject"].values():
            if x_object["/Subtype"] == "/Image":
                data = x_object.get_data()
                with open(f"{output_dir}/page{page_number + 1}_img.png", "wb") as f:
                    f.write(data)
                print(f"Saved image for page {page_number + 1}")

# Example usage
pdf_path = '../../data/documents/genesis-dh-manual-ko.pdf'
output_dir = "../../data/documents/output_images"  # Directory to save extracted images
extract_images_with_pypdf2(pdf_path, output_dir)
```

```python
import fitz  # PyMuPDF
import cv2
import numpy as np
from io import BytesIO
import os


def extract_images_with_opencv_and_text(
    pdf_path, output_dir, image_format="png", separation_fix=True
):
    """
    Extract images from a PDF file using OpenCV and save them with page, image index, and text in filenames.
    Args:
        pdf_path (str): Path to the input PDF file.
        output_dir (str): Directory where extracted images will be saved.
        image_format (str): Format to save images (e.g., "png" or "jpg").
        separation_fix (bool): Whether to fix color issues for 'Separation' color space.
    """
    os.makedirs(output_dir, exist_ok=True)
    pdf_document = fitz.open(pdf_path)

    for page_number in range(len(pdf_document)):
        page = pdf_document[page_number]
        print(f"Processing page {page_number + 1}")

        # Extract text from the page
        page_text = page.get_text("text")  # Extract raw text for file naming
        page_text = "".join(
            c if c.isalnum() else "_" for c in page_text.strip()
        )[:50]  # Keep the first 50 characters of sanitized text

        # Get images on the page
        images = page.get_images(full=True)

        for img_index, img in enumerate(images):
            xref = img[0]
            color_space = img[5]  # Color space (e.g., 'Separation', 'ICCBased')

            try:
                # Extract image bytes
                base_image = pdf_document.extract_image(xref)
                image_bytes = base_image["image"]

                # Decode the image bytes using OpenCV
                image_array = np.frombuffer(image_bytes, dtype=np.uint8)
                image = cv2.imdecode(image_array, cv2.IMREAD_UNCHANGED)  # Load as is

                # Fix 'Separation' color space if needed
                if separation_fix and color_space == "Separation":
                    if len(image.shape) == 2:  # Grayscale
                        image = cv2.cvtColor(image, cv2.COLOR_GRAY2RGB)
                    elif image.shape[2] == 4:  # RGBA
                        image = cv2.cvtColor(image, cv2.COLOR_BGRA2BGR)
                    elif image.shape[2] == 3:  # Already RGB
                        pass

                # Generate output file name
                output_filename = f"page{page_number+1}_img{img_index+1}_{page_text}.{image_format}"
                output_path = os.path.join(output_dir, output_filename)

                # Save the image
                cv2.imwrite(output_path, image)
                print(f"Saved image: {output_path}")

            except Exception as e:
                print(
                    f"Error processing image on page {page_number + 1}, image {img_index + 1}: {e}"
                )


# Example usage
pdf_path = '../../data/documents/genesis-dh-manual-ko.pdf'
output_dir = "../../data/documents/output_images"  # Directory to save extracted images
image_format = "png"  # Format to save images (e.g., "png" or "jpg")

extract_images_with_opencv_and_text(pdf_path, output_dir, image_format)
```

```python
import fitz  # PyMuPDF
import cv2
import numpy as np
from io import BytesIO
import os
import re


def extract_images_with_key_text(
    pdf_path, output_dir, image_format="png", key_pattern=r"[A-Z0-9]{6,}"
):
    """
    Extract images from a PDF and save them with a specific key text value in filenames.
    Args:
        pdf_path (str): Path to the input PDF file.
        output_dir (str): Directory where extracted images will be saved.
        image_format (str): Format to save images (e.g., "png" or "jpg").
        key_pattern (str): Regular expression pattern to extract key text values.
    """
    os.makedirs(output_dir, exist_ok=True)
    pdf_document = fitz.open(pdf_path)

    # Compile regex pattern for key text extraction
    key_regex = re.compile(key_pattern)

    for page_number in range(len(pdf_document)):
        page = pdf_document[page_number]
        print(f"Processing page {page_number + 1}")

        # Extract text blocks and their bounding boxes
        text_blocks = page.get_text("blocks")  # List of (x0, y0, x1, y1, "text")

        # Get images on the page
        images = page.get_images(full=True)

        for img_index, img in enumerate(images):
            xref = img[0]
            img_bbox = page.get_image_bbox(xref)  # Image bounding box

            # Find key text nearest to the image
            key_text = "unknown"
            min_distance = float("inf")
            for block in text_blocks:
                block_bbox = block[:4]  # (x0, y0, x1, y1)
                text = block[4]

                # Extract key values from text using regex
                keys = key_regex.findall(text)
                if keys:
                    # Calculate distance between image and text block
                    distance = bbox_distance(img_bbox, block_bbox)
                    if distance < min_distance:
                        min_distance = distance
                        key_text = keys[0]  # Use the first matching key
                       

            try:
                # Extract image bytes
                base_image = pdf_document.extract_image(xref)
                image_bytes = base_image["image"]

                # Decode the image bytes using OpenCV
                image_array = np.frombuffer(image_bytes, dtype=np.uint8)
                image = cv2.imdecode(image_array, cv2.IMREAD_UNCHANGED)  # Load as is

                # Save the image with the key text in the filename
                
                output_filename = f"page{page_number+1}_img{img_index+1}_{key_text}.{image_format}"
                output_path = os.path.join(output_dir, output_filename)

                cv2.imwrite(output_path, image)
                print(f"Saved image: {output_path}")

            except Exception as e:
                print(
                    f"Error processing image on page {page_number + 1}, image {img_index + 1}: {e}"
                )


def bbox_distance(bbox1, bbox2):
    """
    Calculate the distance between two bounding boxes.
    """
    x1, y1, x2, y2 = bbox1
    bx1, by1, bx2, by2 = bbox2

    # Center points of the bounding boxes
    center1 = ((x1 + x2) / 2, (y1 + y2) / 2)
    center2 = ((bx1 + bx2) / 2, (by1 + by2) / 2)

    # Euclidean distance between centers
    return ((center1[0] - center2[0]) ** 2 + (center1[1] - center2[1]) ** 2) ** 0.5


# Example usage
pdf_path = '../../data/documents/genesis-dh-manual-ko.pdf'
output_dir = "../../data/documents/output_images"  # Directory to save extracted images
image_format = "png"  # Format to save images (e.g., "png" or "jpg")
key_pattern = r"[A-Z0-9]{6,}"  # Regex pattern to match key text

extract_images_with_key_text(pdf_path, output_dir, image_format, key_pattern)
```

```python
import fitz  # pymupdf의 핵심 모듈

def extract_and_chunk_pdf(pdf_path, output_path, chunk_size=500):
    # PDF 열기
    pdf_document = fitz.open(pdf_path)
    chunks = []
    
    for page_number in range(len(pdf_document)):
        # 페이지 읽기
        page = pdf_document[page_number]
        text = page.get_text()
        
        # 페이지별 텍스트를 청킹
        for i in range(0, len(text), chunk_size):
            chunk_text = text[i:i + chunk_size]
            chunks.append({
                "page": page_number + 1,  # 페이지 번호 기록
                "chunk": chunk_text.strip()
            })
    
    # 마크다운 형식으로 저장
    with open(output_path, 'w', encoding='utf-8') as md_file:
        for idx, chunk in enumerate(chunks, start=1):
            md_file.write(f"### Chunk {idx} (Page {chunk['page']})\n")
            md_file.write(chunk["chunk"] + "\n\n")
            md_file.write("---\n\n")  # 청크 구분자
    
    print(f"Chunked Markdown file saved at {output_path}")

# 사용 예제
pdf_path = "../../data/documents/genesis-dh-manual-ko.pdf"  # PDF 파일 경로
output_path = "output_chunks.md"  # 저장할 마크다운 파일 경로
extract_and_chunk_pdf(pdf_path, output_path, chunk_size=500)
```
