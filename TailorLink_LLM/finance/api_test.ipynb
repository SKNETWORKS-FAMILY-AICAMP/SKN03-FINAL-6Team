```python
import base64
import requests
import urllib.parse

def publish_token(client_id, client_secret):
    try:
        # URL 및 요청 파라미터 설정
        oauth_domain = "https://oauth.codef.io"  # 실제 OAuth 도메인으로 변경
        get_token_endpoint = "/oauth/token"
        url = f"{oauth_domain}{get_token_endpoint}"
        params = "grant_type=client_credentials&scope=read"

        # 클라이언트 아이디와 시크릿코드 Base64 인코딩
        auth = f"{client_id}:{client_secret}"
        auth_encoded = base64.b64encode(auth.encode()).decode()  # Base64 인코딩
        auth_header = f"Basic {auth_encoded}"

        # 요청 헤더 설정
        headers = {
            "Content-Type": "application/x-www-form-urlencoded",
            "Authorization": auth_header
        }

        # POST 요청 전송
        response = requests.post(url, headers=headers, data=params)

        # 응답 코드 확인
        if response.status_code == 200:
            # JSON 응답 처리
            response_data = response.json()
            return response_data
        else:
            print(f"Error: HTTP {response.status_code}")
            print(response.text)
            return None

    except Exception as e:
        print(f"Exception occurred: {e}")
        return None
```

```python
# 테스트용 실행 코드
if __name__ == "__main__":
    client_id = client_id       # 클라이언트 ID 입력
    client_secret = client_secret  # 클라이언트 시크릿 입력
    token_response = publish_token(client_id, client_secret)
    if token_response:
        print("Access Token Response:")
        print(token_response)
    else:
        print("Failed to get access token.")
```
