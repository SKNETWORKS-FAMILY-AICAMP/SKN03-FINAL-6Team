```python
from cryptography.hazmat.primitives.asymmetric import rsa
from cryptography.hazmat.primitives.asymmetric import padding
from cryptography.hazmat.primitives import hashes

# RSA 키 생성
private_key = rsa.generate_private_key(public_exponent=65537, key_size=2048)
public_key = private_key.public_key()

# 암호화할 주민등록번호
rrn = "1054411".encode()

# 공개키로 암호화
encrypted_rrn = public_key.encrypt(
    rrn,
    padding.OAEP(
        mgf=padding.MGF1(algorithm=hashes.SHA256()),
        algorithm=hashes.SHA256(),
        label=None
    )
)
print("Encrypted RRN:", encrypted_rrn)
```

```python
# RSA 암호화된 데이터를 Base64로 인코딩
encrypted_rrn_base64 = base64.b64encode(encrypted_rrn).decode('utf-8')

# Base64 데이터를 URL 인코딩
encrypted_rrn_url_encoded = urllib.parse.quote(encrypted_rrn_base64)
```

```python
# URL 인코딩
rrn_url_encoded = urllib.parse.quote(encrypted_rrn)

print("URL Encoded RRN:", rrn_url_encoded)
```
